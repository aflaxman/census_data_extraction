# Visualizing Census Data in Python

In a [previous post](http://andrewgaidus.com/Reading_Zipped_Shapefiles/) I demonstrated how to read geographic data from a website URL (in the form of Shapefiles) directly into Python. In this post, I will use this tool in conjunction with the Census API to map demographic data. The Census API provides access to thousands of different demographic variables at all Census geographic levels going back in time. Ultimately, I will use these tools to create a set of dot-density maps that provide an easy way to visualize population distributions within a county by race and ethnicity. 

### Dot Density Maps
[Dot-density maps](https://en.wikipedia.org/wiki/Dot_distribution_map) show concentrations of subjects within a given area. If we have counts of population within block groups within a county, the county population distribution can be modeled by randomly distributing points within each block group, using a fixed ratio of population per point. These maps sometimes work better than [choropleth maps](https://en.wikipedia.org/wiki/Choropleth_map) for presenting data that varies continuously over space (such as population distributions).

I've written a Python module called ```census mapper``` that contains several functions that I use throughout this post for (1) extracting census geographic and demographic data, (2) implementing the random point in polygon algorithm needed to create a dot-density map, and (3) actually plotting the data. The [code](https://github.com/agaidus/census_data_extraction/blob/master/census_mapper.py) for these functions can be found on my [GitHub](https://github.com/agaidus/census_data_extraction). 

Note that I will be working with the Census data at the block group level, because this is the most finely-grained geography for which demographic data is available. As a size reference, block groups are defined to have between 600 and 3,000 people. Also, I will be using the [2015 American Community Survey](https://www.census.gov/programs-surveys/acs/) (ACS) 5-Year Averages as my Census Demographic data source. The ACS is an ongoing survey that provides data estimates between the Decennial Census. The 2015 5-Year Averages are the most current data that can be obtained at the block group level. 

In general I take the following steps, which use the following functions that I developed:

1. Read in a county's block group boundaries from the Census FTP site (```zip_shp_to_gdf```)

2. Obtain a county's 2015 ACS block group demographic data from the Censys API (```get_census_variables```)

3. Take a population group of interest and randomly generate points within block groups across the county (```gen_points_in_gdf_polys```)

4. Pull all of this data together and map it! (```gen_count_dot_density_map```)

Before actually making the dot density maps, I demonstrate use of the two functions used to extract Census geography and data and show how these can be realy easily used to make a quick map.

````python
from census_mapper import zip_shp_to_gdf, get_census_variables, gen_points_in_gdf_polys
from shapely.geometry import Polygon
import matplotlib.pyplot as plt
import geopandas as gpd
import pandas as pd
%matplotlib inline
````

## 1 - Read in Block Group Geography within a County
The details of the code behind this function are described in an earlier [post](http://andrewgaidus.com/Reading_Zipped_Shapefiles/), but here I just demonstrate its use with a quick example. I provide the Census FTP address for the Shapefile of California Census Tracts and I return a GeoDataFrame. I subset this GeoDataFrame to include only Alameda County Tracts and then reproject it to a new coordinate reference system.

````python
zipped_shp_url = 'http://www2.census.gov/geo/tiger/GENZ2015/shp/cb_2015_06_tract_500k.zip'

all_ca_tract_geo = zip_shp_to_gdf(zipped_shp_url)
alameda_tract_geo = all_ca_tract_geo[all_ca_tract_geo.COUNTYFP=='001']
alameda_tract_geo = alameda_tract_geo.set_index("GEOID")['geometry'].to_crs(epsg=3310)
print alameda_tract_geo.head()
````

## 2 - Get ACS Demographic Data
Now that I have my geography, I need some demographic data to match it to. While I ultimately will be using population counts, in this example I extract the variable for Median Household Income. I specify the specific dataset I want (2015, ACS, Census Tracts); the geographic extent of my data (Alameda County, which is FIPS 06001); and the set of variables that I'm interested in (median household income, which is ACS variable B19013_001E. The full set of variable names and descriptions available from the census can be found [here](http://api.census.gov/data/2015/acs5/variables.html). 

This function then returns a DataFrame indexed by Tract FIPS that contains my specified variables. Note that I also specified an optional variable label parameter, which is just used to rename it to something more legible.

````python
alameda_tract_income = get_census_variables(2015, 'acs5', 'tract', 
                                            {'county':'001', 'state':'06'}, 
                                            ['B19013_001E'], ['income'])['income'].dropna().astype(int)
alameda_tract_income.head()
````

Now that I have my data and geography I can really easily make a quick choropleth map of income level using the ```GeoPandas``` built in plotting functions. If you are familiar with the Bay Area, you'll see that the map shows a pattern we might expect - wealthier people live in the Oakland / Berkeley Hills and the suburbs.

````python
alameda_gdf = gpd.GeoDataFrame(geometry = alameda_tract_geo, data = alameda_tract_income)
fig, ax = plt.subplots(figsize=(10,10))
ax.set(aspect='equal', xticks=[], yticks=[])
alameda_gdf.plot(column= 'income', ax = ax, scheme='QUANTILES', cmap='Purples', legend=True)
plt.title('Alameda County, CA - Median Household Income', size = 14)
````

## 3 -  Randomly Generating Points within Polygons

My ```census_mapper``` script contains a function ```called gen_points_in_gdf_polys```, which is used to create the points that will be plotted in the dot-density maps. Below I demonstrate its functionality with a simple example. Let's say I have the following 3 polygons in my dataset, each of which has a different (specified) number of people living in it. 

````python
p1 = Polygon([[0,2],[1,3],[2,2]])
p2 = Polygon([[0,0],[0,2],[2,2],[2,0]])
p3 = Polygon([[2,0],[2,3],[3,3],[3,0]])
g = gpd.GeoSeries([p1,p2,p3])

vals = pd.Series([100, 50, 500])

ax = g.plot()
for label,loc in zip(vals, g.centroid):
    ax.text(loc.x-.2, loc.y, str(label), fontsize=15)
````

My function randomly generates a point within the minimum bounds of a polygon, checks if it actually falls within the polygon, and then adds it to a list if it does. This is done until the specified number of points is generated. This function is then applied over the entire dataset of Polygons and values.

Below, I plot the output of this function, which makes it really easy to compare densities across those polygons. In creating the dot density maps, this same function is applied to all block groups within a county.

````python
random_points = gen_points_in_gdf_polys(geometry = g, values=vals)

ax = g.plot()
random_points.plot(color='black', ax=ax, alpha=.7)
````

## 4 - Create Dot Density Maps

Now that we have our tools to extract the data and generate the dot-density points, we're ready to create some maps with real data. The function ```gen_count_dot_density_map``` wraps the other functions I defined above. The only required input is a 

don't show bg borders
allows for specificiation of one field or multiple fields

````python
def gen_count_dot_density_map(county_fips, plot_field=None, mult_fields = None,
                                pts_per_person = 400, place_labels = True, 
                                label_dot_cutoff = 20, epsg = 2163, seed=10,
                                dot_color = 'black', dot_transparency=0.4, 
                                figsize=(12,12), ax=None, title=None):

    #get geodataframe of block group shapefile
    bgfile_name = 'http://www2.census.gov/geo/tiger/GENZ2015/shp/cb_2015_{}_bg_500k.zip'.format(county_fips[:2])
    bg_geo = zip_shp_to_gdf(bgfile_name)
    #subset to those that are in the county and project it to the CRS
    bg_geo=bg_geo[bg_geo['GEOID'].str[:5]==county_fips].to_crs(epsg=epsg).set_index("GEOID")['geometry']
    
    #specify variable list and variable names for the census api function
    varlist = ['B02001_001E', 'B03002_003E', 'B03002_012E', 'B02001_002E', 
               'B02001_003E', 'B02001_005E', 'B02001_004E', 'B02001_006E',
               'B02001_007E', 'B02001_008E']
    names = ['total', 'white_nhs', 'hispanic', 'white',
             'black', 'asian', 'ai_an', 'nh_pi', 
             'other', 'two_plus']
    #read in block group level census variables
    dems = get_census_variables(2015, 'acs5', 'block group', 
                                {'county':county_fips[2:], 
                                 'state':county_fips[:2]}, varlist, names)
    
    #Calculate the union of block groups as the county boundaries
    union = gpd.GeoSeries(bg_geo.unary_union)
    
    #if axes object is specified, plot to this axis, otherwise create a new one
    if ax:
        union.plot(color='white', figsize=figsize, ax=ax)
    else:
        ax = union.plot(color='white', figsize=figsize)
    #set aspect equal and add title if specified
    ax.set(aspect='equal', xticks=[], yticks=[])
    if title: ax.set_title(title, size=15)
    
    
    #For plotting one (rather than multiple) fields
    #Use function to generate points randomly within bg boundaries
    if plot_field:
        ps = gen_points_in_gdf_polys(geometry = bg_geo, values=dems[plot_field], 
                               points_per_value = pts_per_person, seed=seed)
        ps.plot(ax=ax, markersize=2, color=dot_color, alpha=dot_transparency)
        
    #For plotting multiple fields
    #Same thing, but now generate one geodataframe with points labeled by the field they are symbolizing   
    if mult_fields:
        list_of_point_categories = []
        for field in mult_fields:
            ps=gpd.GeoDataFrame(gen_points_in_gdf_polys(geometry = bg_geo, values=dems[field],
                                 points_per_value = pts_per_person, seed=seed))
            ps['field']=field
            list_of_point_categories.append(ps)
        all_points=gpd.GeoDataFrame(pd.concat(list_of_point_categories))
        all_points.plot(ax=ax, markersize=2, alpha=dot_transparency, 
                  column='field', categorical=True, legend=True)
    # if place labels requested, read in place shapefile
    # determine the number of dots in each place to have a threshold for whether they should be labeled or not
    if place_labels:
        place_name = 'https://www2.census.gov/geo/tiger/TIGER2010/PLACE/2010/tl_2010_{}_place10.zip'.format(county_fips[:2])
        place_geo = zip_shp_to_gdf(place_name)[['NAME10','geometry']].to_crs(epsg=epsg)
        
        # determine the number of dots in each place to have a threshold for whether they should be labeled or not
        totals_ps = gen_points_in_gdf_polys(geometry = bg_geo, values=dems['total'], 
                           points_per_value = pts_per_person, seed=seed)
        points_to_join = gpd.GeoDataFrame(totals_ps.reset_index())
        points_to_join.crs = place_geo.crs
        city_dots = gpd.tools.sjoin(points_to_join, place_geo).groupby('NAME10').size()
        #include those labels that have more than the specified number of dots per place
        include = city_dots[city_dots>=label_dot_cutoff].index
        
        # get centroid of places to be used in label location
        place_geo = gpd.GeoDataFrame(place_geo, geometry = place_geo.centroid)
        place_geo = place_geo[(place_geo.within(union.iloc[0]))&place_geo.NAME10.isin(include)]
        labels = place_geo.apply(lambda z:ax.annotate(s=z.NAME10, xy=z.geometry.coords[0],
                                                      ha='center', fontsize=14, fontweight='bold', alpha=0.9, 
                                                      fontname="Palatino Linotype"), axis=1)
    return ax
````

````python
gen_count_dot_density_map('06001', 'total', title='Alameda County, CA Population Distribution', dot_transparency=.25, 
                           label_dot_cutoff = 250, pts_per_person=200, dot_color='red')
````

````python
def compare_pop_distributions(county_fips, name=None):
    races = ['white_nhs', 'hispanic', 'black', 'asian']
    race_labels = ['White', 'Hispanic','Black','Asian']
    fig, axes = plt.subplots(2,2)
    for race, race_label, ax in zip(races, race_labels, fig.axes):
        ax=gen_count_dot_density_map(county_fips, race, place_labels=False, 
                       ax=ax, title=race_label)
    
    
    fig.tight_layout()
    bbox = ax.get_window_extent().transformed(fig.dpi_scale_trans.inverted())
    width, height = bbox.width, bbox.height
    figwidth, figheight = 12, 12*(height/width)
    #fig.set_size_inches(figwidth, figwidth)
    #plt.subplots_adjust(wspace=.7/figwidth, hspace = .7/figheight, top=0.92)
    #fig.suptitle(name, size=14)

compare_pop_distributions('36061', 'Manhattan, NY')
````
