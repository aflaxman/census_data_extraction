```python
from census_mapper import zip_shp_to_gdf, get_census_data
import matplotlib.pyplot as plt
import geopandas as gpd
%matplotlib inline
```

```python
#zipped_shp = 'http://www2.census.gov/geo/tiger/TIGER2010/BG/2010/tl_2010_06001_bg10.zip'
zipped_shp = 'http://www2.census.gov/geo/tiger/GENZ2015/shp/cb_2015_06_tract_500k.zip'
gdf = zip_shp_to_gdf(zipped_shp).to_crs(epsg=3310)
```

```python
subgdf = gdf[gdf.COUNTYFP=='001']
```

```python
fig, ax = plt.subplots(figsize=(20,20))
ax.set(aspect='equal', xticks=[], yticks=[])
subgdf.plot(color = 'white', ax = ax)
plt.title('Alameda County, CA', size=18)
```

[variables](http://api.census.gov/data/2014/acs1/variables.html)

```python
var_list = ['B19013_001E']
bar_labels = ['income']

df = get_census_data(2015, 'acs5', 'tract', {'county':'001', 'state':'06'}, ['B19013_001E'], ['income'])
income = gpd.GeoDataFrame(geometry = subgdf.set_index('GEOID')['geometry'], data = df['income'].dropna().astype(int))
```

```python
fig, ax = plt.subplots(figsize=(20,20))
ax.set(aspect='equal', xticks=[], yticks=[])
income.plot(column= 'income', ax = ax, scheme='QUANTILES', cmap='Purples', legend=True)
plt.title('Alameda County, CA', size = 18)
```
